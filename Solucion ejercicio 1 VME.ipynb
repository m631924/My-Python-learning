```python
import numpy as np

'''
               [Fav, Desfavorable]
Planta grande  [100,  10]
Planta pequeña [60, 30]
No hacer nada  [0,  0]
'''
viEstados = 2
viAlternativas = 3
vsEstados = ["Favorable", "Desfavorable"]
vsAlternativas = ["Planta grande", "Planta pequeña", "No hacer nada"]
# Datos matriz de pagos
vlstMatrizPagos = [[100, -100],[60,30],[0,0]]
vlstProb = [0.6, 0.4]

# ejercicio 1 VME
vlstAlternativaEval = []
vrValorEsperadoEval = 0
vrMejorValorEsperado = 0
vsMejorAlternativa = ""
vsAlternativaEval = ""
vlstVME = []

for iLoop in range(viAlternativas):
  vlstAlternativaEval = vlstMatrizPagos[iLoop]
  vrValorEsperadoEval= (sum(np.multiply(vlstAlternativaEval, vlstProb)))
  vsAlternativaEval = vsAlternativas[iLoop]
  vlstVME.append(vrValorEsperadoEval)
  print(vrValorEsperadoEval)
  if iLoop == 0:
    # inicializar vrMejorValorEsperado y vsMejorAlternativa
    vrMejorValorEsperado = vrValorEsperadoEval
    vsMejorAlternativa = vsAlternativaEval

  if  vrValorEsperadoEval > vrMejorValorEsperado:
    vrMejorValorEsperado = vrValorEsperadoEval
    vsMejorAlternativa = vsAlternativaEval

print('La mejor alternativa es '+ vsMejorAlternativa)
```
