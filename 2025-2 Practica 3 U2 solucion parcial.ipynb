## Algoritmo de Prim

El siguiente codigo contiene un ejemplo de resolución de un arbol de expansion minima usando el algoritmo de Prim.

````python
import networkx as nx
import matplotlib.pyplot as plt
import random
infinito = float("inf")
vlstNodos = ["A", "B", "C", "D", "E", "F", "G"]
vlstNodosNoConectados = ["A", "B", "C", "D", "E", "F", "G"]
vlstNodosConectados = []
random.seed(25)

# Datos: Lista de adyacencia
# Diccionario Nodoi : [("Nodoj", distanciaij)]
vlistaAdyacencia = {"A" : [("B", 7), ("D", 5)],
           "B" : [("A", 7), ("C", 8), ("D", 9), ("E", "7")],
           "C" : [("B", 8), ("E", 5)],
           "D" : [("A", 5), ("B", 9), ("E", 15), ("F", 6)],
           "E" : [("B", 7), ("C", 5), ("D", 15), ("F", 8), ("G", 9)],
           "F" : [("D", 6), ("E", 8), ("G", 11)],
           "G" : [("E", 8), ("F", 11)]
}

# Método para dibujar grafo con etiquetas
def mbDibujarGrafo(poGrafo):
  pos = nx.spring_layout(poGrafo, seed = 100) # Evitar que cambie la posición de los nodos
  nx.draw(poGrafo, pos = pos, with_labels = True)
  # Añadir etiqueta de cada arco
  labels = nx.get_edge_attributes(poGrafo,'weight') # obtiene datos del grafo
  nx.draw_networkx_edge_labels(poGrafo, pos, edge_labels=labels)

  # Colorear solución : arcos o aristas
  set1 = vlstTrayectoriaSolucion
  colors = nx.get_edge_attributes(poGrafo,'color').values()
  for i in [set1]:
    nx.draw_networkx_edges(G,pos, edgelist = i, edge_color=colors)

  plt.show()

# Método para obtener lista de nodos vecinos
def mlstNodosVecinos(poGrafo = "", psNodo=""):
  lstNodos = []
  if psNodo=="":
    print("Nodo no puede ser vacio, se retornara lista vacia")
    return lstNodos
  if poGrafo == "":
    lstNodosDist = vlistaAdyacencia[psNodo]
    for iNodo in lstNodosDist:
      lstNodos.append(iNodo[0]) # Obtener nombre del nodo
    return lstNodos
  else:
    # o usar el método de la librería Networkx:
    return(list(poGrafo.neighbors(psNodo)))

# Método para obtener lista de nodos vecinos y sus distancias
def mlstNodosVecinosPesos(poGrafo, psNodo):
  # nodo origen y destino
  sNodoOrigen = psNodo
  sNodoDestino = ""
  lstNodosDestino = mlstNodosVecinos(poGrafo, psNodo)
  rPeso = 0
  lstNodosPeso =[]
  # obtener diccionario de arcos y pesos
  dictValores = nx.get_edge_attributes(poGrafo, "weight")
  for iLoop in range (len(lstNodosDestino)):
    sNodoDestino = lstNodosDestino[iLoop]
    sLlave = (sNodoOrigen, sNodoDestino)
    if sLlave not in dictValores:
      #cambiar sentido de la llave
      ##llave no existe en este sentido u->V, puede que exista de v->u (Grafo No Dirigido)
      sLlave = (sNodoDestino, sNodoOrigen)
    rPeso = dictValores[sLlave]
    lstNodosPeso.append([sNodoDestino, rPeso])

  return lstNodosPeso

# Método para separar listas dentro de listas, devuelve dos listas
def mlstSplitLista(plstLista):
  # funciona solo para 2 elementos de la sublista
  # asumiendo que la lista grande contiene sublistas con los pares [nodo, peso]
  lstLista1 = []
  lstLista2 = []

  for elemento in plstLista:
    lstLista1.append(elemento[0]) # agregar Nodo
    lstLista2.append(elemento[1]) # agregar peso
  return lstLista1, lstLista2

# Método para agregar un nuevo atributo a todos los nodos
def mbNuevoAtributoNodo(poGrafo, psNombreAtributo, paValor):
  nx.set_node_attributes(poGrafo, paValor, psNombreAtributo)

# Método para marcar un nodo como conectado
def mbConectarNodo(poGrafo, psNodo):
  # Actualizar lista de nodos conectados y desconectados
  vlstNodosNoConectados.remove(psNodo) # Retirar nodo
  vlstNodosConectados.append(psNodo) # Agregar nodo
  # Actualizar attributo "conectado" en el grafo
  poGrafo.nodes[psNodo]["conectado"] = True

# Método para cambiar color de una arista, por defecto rojo
def mbCambiarColorArista(poGrafo, psNodoOrigen, psNodoDestino, psColor="r"):
  poGrafo.edges[psNodoOrigen, psNodoDestino]["color"] = psColor


# Definir/Crear el grafo a partir de los datos (lista de adyacencia)
G = nx.Graph()
for iLoop1 in range(len(vlstNodos)):
  sNodoOrigen = vlstNodos[iLoop1]
  lstConexionesNodo = vlistaAdyacencia[sNodoOrigen]
  for iLoop2 in range(len(lstConexionesNodo)):
    sNodoDestino = lstConexionesNodo[iLoop2][0]
    iPesoArco = lstConexionesNodo[iLoop2][1]
    # Añade arco por arco
    G.add_edge(sNodoOrigen, sNodoDestino, weight = iPesoArco)

# Agregar atributos "conectado" a toods los nodos (atributo de control)
mbNuevoAtributoNodo(G, "conectado", False)


## Algoritmo de Prim
iCount = 0
vlstTrayectoriaSolucion = []
vlstCostoSolucion = []
while len(vlstNodosNoConectados) > 0:
  rPeso = 0
  vrMejorPeso = 0
  iCount += 1
  #print("iteración.." + str(iCount))
  # detener bucle por seguridad
  if iCount > 1000:
    break
  # solo iteración 1
  if len(vlstNodosConectados) == 0:
    vsNodoInicio = random.choice(vlstNodos)
    mbConectarNodo(G, vsNodoInicio)

  # Encontrar mejor nodo con el que conectarse
  rNodoPesoEval = 0
  rNodoPesoMenor = 0
  sNodoPesoMenor = ""
  lstParcialPesos = []
  lstParcialNodosOrigen = []
  lstParcialNodosDestino = []
  # lista de nodos conectados y sus posibles arcos
  for iLoop in range(len(vlstNodosConectados)):
    sNodoOrigen = vlstNodosConectados[iLoop]
    lstVecinosPesos = mlstNodosVecinosPesos(G, sNodoOrigen) # retorna [["nodo", peso], ["nodo2", peso2],...]
    lstSearchNodos, lstSearchPesos=mlstSplitLista(lstVecinosPesos)
    # iterar en la lista de pesos, no necesariamente el nodo con menor peso está disponible
    for iLoop2 in range(len(lstSearchNodos)):
      sNodoDestino = lstSearchNodos[iLoop2]
      rPeso = lstSearchPesos[iLoop2]
      if G.nodes[sNodoDestino]["conectado"] == False:
        # nodo no conectado aun, evaluar si es el menor
        if iLoop2 == 0:
          # resetear valores iniciales para evitar errores
          sNodoPesoMenor = sNodoDestino
          rNodoPesoMenor = rPeso

        if (rPeso < rNodoPesoMenor):
          rNodoPesoMenor = rPeso
          sNodoPesoMenor = sNodoDestino
        # Agregar a lista parcial de mejores pesos y nodos
        lstParcialNodosOrigen.append(sNodoOrigen)
        lstParcialNodosDestino.append(sNodoDestino)
        lstParcialPesos.append(rPeso)
  # Seleccionar el mejor de los mejores pesos
  vrMejorPeso = min(lstParcialPesos)
  viIndiceMejorPeso = lstParcialPesos.index(vrMejorPeso)
  sNodoOrigen = lstParcialNodosOrigen[viIndiceMejorPeso]
  sNodoDestino = lstParcialNodosDestino[viIndiceMejorPeso]
  rPeso = vrMejorPeso
  # Nodo origen ya esta conectado,
  # marcar nodo destino como conectado
  mbConectarNodo(G, sNodoDestino)

  # Almacenar trayectoria de la solucion
  vlstTrayectoriaSolucion.append((sNodoOrigen, sNodoDestino))
  vlstCostoSolucion.append(rPeso)
print("Los arcos conectados son : " + str(vlstTrayectoriaSolucion))
print("La distancia total recorrida es de "+ str(sum(vlstCostoSolucion)))


# Aplicar colores a todas las aristas de la solución
for iLoop in range(len(vlstTrayectoriaSolucion)):
  sArista = vlstTrayectoriaSolucion[iLoop]
  if sArista in G.edges:
    sElem1 = sArista[0]
    sElem2 = sArista[1]
    mbCambiarColorArista(G, sElem1, sElem2, "red")
  else:
    # se encuentra en el sentido contrario
    sElem1 = sArista[0]
    sElem2 = sArista[1]
    mbCambiarColorArista(G, sElem2, sElem1, "red")


mbDibujarGrafo(G)


````

# Minimum spanning tree (Árbol de expansión mínima)

**minimum_spanning_tree(G, weight='weight', algorithm='kruskal', ignore_nan=False)**

Returns a minimum spanning tree or forest on an undirected graph G

**Parameters:**
* G: undirected graph
  An undirected graph. If G is connected, then the algorithm finds a spanning tree. Otherwise, a spanning forest is found.
* weight : str
  Data key to use for edge weights.
* algorithm : string
  The algorithm to use when finding a minimum spanning tree.
  Valid choices are ‘kruskal’, ‘prim’, or ‘boruvka’. The default is ‘kruskal’.
* ignore_nan
bool (default: False)
If a NaN is found as an edge weight normally an exception is raised. If ignore_nan is True then that edge is ignored instead.

````python
import networkx as nx
import matplotlib.pyplot as plt

# input
vData = [(1, 2, 224), (1, 6, 300), (1, 7, 539),
         (2, 1, 224), (2, 3, 200), (2, 7, 539),
         (3, 2, 200), (3, 4, 400), (3, 7, 600),
         (4, 3, 400), (4, 5, 400), (4, 7, 200),
         (5, 4, 400), (5, 6, 600), (5, 7, 447),
         (6, 1, 300), (6, 5, 600), (6, 7, 283),
         (7, 1, 539), (7, 2, 539), (7, 3, 600), (7, 4, 200), (7, 5, 447), (7, 6, 283)]

fixed_positions = {1: (1,2.5),
                   2: (2,4),
                   3: (4,4),
                   4: (5,2.5),
                   5: (4,1),
                   6: (2,1),
                   7: (3,2.5)
                   }

# Aplicar algoritmo y almacenar arbol resultante en una variable
G = nx.Graph()
G.add_weighted_edges_from(vData)
# opción 1, usar algoritmo por defecto (Kruskal)
#T = nx.minimum_spanning_tree(G)
# opción 2, usar algoritmo de Prim
T = nx.minimum_spanning_tree(G, weight='weight', algorithm='prim')
# opción 3, usar algoritmo de Kruskal
#T = nx.minimum_spanning_tree(G, weight='weight', algorithm='kruskal')


# output
sorted(T.edges(data=True))
vlstAristasMST = T.edges()
# calcular el peso total de las aristas que componen el árbol de expansión míninma
vrSumaPeso = 0
for arco in vlstAristasMST:
  node1 = arco[0]
  node2 = arco[1]
  # obtener peso de cada artista y hacer sumatoria
  vDiccionarioPeso = T.get_edge_data(node1, node2, default = 'weight')
  vrPeso = vDiccionarioPeso['weight']
  vrSumaPeso += vrPeso
# mostrar resultados en letras (Resumen)
print('Solución: las aristas del árbol de expansión mínima son: ', vlstAristasMST)
print('El peso total del del árbol de expansión mínima es', vrSumaPeso)

# Colorear solución : arcos o aristas
for arco in vlstAristasMST:
  node1 = arco[0]
  node2 = arco[1]
  G.edges[node1, node2]['color'] = 'red'
  #G.edges[node2, node1]['color'] = 'red' # /!\ Descomentar/probar con grafos dirigidos

fixed_nodes = fixed_positions.keys()
labels = nx.get_edge_attributes(G,'weight') # obtiene datos del grafo
pos = nx.spring_layout(G,pos=fixed_positions, fixed = fixed_nodes)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
colors = nx.get_edge_attributes(G,'color').values()

for arco in vlstAristasMST:
  #print(arco)
  nx.draw_networkx_edges(G,pos, edgelist = [arco], edge_color=colors)

plt.show()


````
