```python
# Fibonacci
def fib (n):
  global numcalls
  #print("fib called with "+ str(n))
  numcalls += 1
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

# Llamar a la funcion Fibonacci
numcalls = 0
print(fib(30))
print("numcalls =" + str(numcalls))
```

```python
# Fibonacci con memoizacion

def fibopt(n, memo):
  global numcalls
  numcalls += 1
  if not n in memo:
    memo[n]=fibopt(n-1, memo) + fibopt(n-2, memo)
    return memo[n]
  else:
    return memo[n]
def fib2(n):
  memo = {0:0, 1: 1}
  fibopt(n, memo)
  return memo[n]

# Llamar a la funcion Fibonacci optimizada
numcalls = 0
print(fib2(30))
print("numcalls =" + str(numcalls))
```

```python
# PD diligencia (cazafortunas)
# 4 etapas
N = 4
s = [["A"], ["B", "C", "D"], ["E", "F", "G"], ["H", "I"]]
x = [["B", "C", "D"], ["E", "F", "G"], ["H", "I"], ["J"]]
c = [[[2, 4, 3]], [[7, 4, 6], [3, 2, 4], [4, 1, 5]],[[1, 4], [6, 3], [3, 3]], [[3], [4]]]

def PDdiligencia(n, s, x, c, foptim = []):
  print("-----------------------")
  print("etapa" +  str(n))
  sn = s[n-1]
  print("estados" + str(sn))
  xn = x[n-1]
  print("variables" + str(xn))
  cn = c[n-1]
  nbs = len(s)
  nbx = len(x[n-1])
  #f = []
  bCostoFuturo = len(foptim) > 0
  foptim2 = []
  # costo
  for i in range(len(sn)):
    f = []
    for j in range(len(xn)):
      sname = sn[i]
      xname = xn[j]
      c1 = cn[i][j]
      if bCostoFuturo:
        # costo optimo siguiente etapa
        c2 = foptim[j][1]
        c1+=c2
      f.append(c1)

    # mejor costo
    cmin = min(f)
    xmin = xn[f.index(cmin)]
    foptim2.append([sname, cmin, xmin])
  if n == 1:
    # terminado
    print(foptim2)
    return(foptim2[0][1])
  else:
    # seguir llamando la funcion
    print(foptim2)
    return PDdiligencia(n-1, s, x, c, foptim2)

 # Llamar algoritmo
PDdiligencia(4, s, x, c)



```
