### FOR

La sentencia for se usa para iterar sobre los elementos de una secuencia (como una cadena de caracteres, tupla o lista) u otro objeto iterable:


````python
print("Primer bucle for")
# variable "i" inicio por defecto en 0 y avanza "n veces"
print("Bucle for 1")
for i in range(10):
    # imprimir los valores que toma i en cada iteración
    print('hola mundo')
    print(i)
print ("Bucle for 2")
for j in range (5, 10):
    # imprimir los valores que toma i en cada iteración
    print('hola mundo 2')
    print (j)
print ("Bucle for 3")
for i in range(10, 5, -1):
    # imprimir los valores que toma i en cada iteración
    print('hola mundo 3')
    print(i)
````

### Detener y/o salir de un bucle FOR o WHILE con BREAK y CONTINUE
Ejecutar el código y observar lo que sucede.

¿En qué se diferencian BREAK y CONTINUE?

````python
# Deteniendo un bucle for usando BREAK y CONTINUE
print("uso de 'break'")
for i in range(10):
    if (i >= 2) and (i<6):
      #interrumpir por completo las iteraciones alcanzada la condición
      break
    # imprimir los valores que toma i en cada iteración
    print(i)
print("uso de 'continue'")
for i in range(10):
    if (i >= 2) and (i<6):
      # saltearse la iteración alcanzada la condición
      continue
    # imprimir los valores que toma i en cada iteración
    print(i)
````

### WHILE
La sentencia while se usa para la ejecución repetida siempre que una expresión sea verdadera

Consideraciones importantes:
1. Si no se llega a la condición deseada, podemos tener un bucle infinito.
2. Evitar usarlo si es posible, viendo otras posibilidades como FOR
3. Tip: usar contador e incrementarlo o disminuirlo por seguridad al llegar a cierto numero. (i += 1)

````python
x = 5
while x > 0:
    x -=1
    print(x)
````

También se pueden anidar bucles WHILE:

````python
# Ejemplo de permutaciones

#inicializacion de variables "i" y "j"
i = 1
j = 1
while i < 4:
    while j < 4:
        print(i,j)
        j += 1
    # incrementar contador
    i += 1
    # reiniciar variable "j"
    j = 1
````

### C. Criterio de realismo (Hurwicz)

Pondera un consenso entre la decisión optimista y pesimista.
1. Para cada alternativa determine el promedio ponderado (PP).
2. Seleccione la alternativa que presenta la máxima recompensa de estas.

````python
# --- Objetivo : buscar el máximo del promedio ponderado

# importar libreria Copy
from copy import deepcopy

# Definir e inicializar variables
viAlpha = 0.80
viBeta = 1 - viAlpha
viValorMinimo = 0
viValorMaximo = 0
viValorActual = 0
viFilaValorMaximo = 0
viColumnaValorMaximo = 0
viPesoPonderado = 0

# Agregar columna "Máximo" a la tabla actual
#Hacer copia profunda de la tabla original
vlstTablaHurwicz = []
vlstTablaHurwicz = deepcopy(vlstTablaRecompensas)


# Recorrer filas
for iRow in range (4):
  if iRow == 0:
    # Agregar columna "Máximo" a la primera fila
    vlstTablaHurwicz[iRow].append("Máximo")
  else:
    # Agregar nueva columna con valor 0 o en blanco
    vlstTablaHurwicz[iRow].append(0)


# Determinar pesos ponderados de cada fila y anotar en columna "Máximo"
for iRow in range(1, 3 + 1):
  # reiniciar variables
  viValorActual = 0
  viPesoPonderado = 0

  for iCol in range (1, 3 + 1):
    # obtener valor con el que ponderar
    viValorActual = vlstTablaHurwicz[iRow][iCol]
    if iCol == 1:
      viPesoPonderado += viValorActual * viAlpha
    elif iCol == 2:
      viPesoPonderado += viValorActual * viBeta
      # Actualizar valor columna "Máximo"
      vlstTablaHurwicz[iRow][3] = viPesoPonderado

# Determinar fila que tiene mayor valor (optimista)

# reiniciar variables
viValorActual = 0
viValorMaximo = 0

for iRow in range (1, 3 + 1):

  viValorActual = vlstTablaHurwicz[iRow][-1]
  if viValorActual >= viValorMaximo:
    # Memorizar valor máximo
    viValorMaximo = viValorActual
    viFilaValorMaximo = iRow


print("El valor máximo que se obtuvo es de " + str(viValorMaximo))

print("Toma de decisión bajo incertidumbre con criterio de realismo : La alternativa que se debe tomar es de '" + vlstTablaHurwicz[viFilaValorMaximo][0] + "'")

#print("Valores de la tabla:")
#print(vlstTablaHurwicz)'''
````

### D. Probabilidades iguales (Laplace)

Utiliza todas las recompensas para cada alternativa.

2. Seleccione la alternativa que presenta la máxima recompensa de estas.

### E. Arrepentimiento (minimax)

Se basa en la pérdida de oportunidad (de no elegir la mejor alternativa).

1. Determine los valores de arrepentimiendo de cada evento.
2. Determine el máximo arrepentimiento para cada alternativa
3. Seleccione la alternativa que presenta el mínimo arrepentimiento
